```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt


#load
rf = pd.read_csv('../outputs/RF_reg_TEST_pred.csv')
lasso = pd.read_csv('../outputs/lasso_reg_TEST_pred.csv')

# check if all rows are the same
if sum(rf['Measured_pH_TEST'] == lasso['Measured_pH_TEST']) == rf.shape[0]:
    print('All rows are the same for the measured pH\n')

# join
test_prep_reg = pd.concat([rf, lasso], axis=1).drop(['Unnamed: 0'],axis=1).iloc[:,0:3]

# save
test_prep_reg.to_csv('../outputs/test_prep_reg.csv')

test_prep_reg.head()
```

```python
_, ax = plt.subplots(dpi=100, ncols = 2, figsize=(12,4) )
nbins=None


sns.regplot(x='Measured_pH_TEST', y = 'Predicted_pH_LASSO_TEST', data = test_prep_reg, x_bins=nbins, ax=ax[0], label='LASSO', ci=90 )
sns.regplot(x='Measured_pH_TEST', y = 'Predicted_pH_RFReg_TEST', data = test_prep_reg, x_bins=nbins, ax=ax[1], label='RF', ci=90)


ax[0].plot(test_prep_reg.Measured_pH_TEST, test_prep_reg.Measured_pH_TEST,'--k')
ax[1].plot(test_prep_reg.Measured_pH_TEST, test_prep_reg.Measured_pH_TEST,'--k')

ax[0].set_title('LASSO \n Test data')
ax[1].set_title('RF \n Test data')

ax[0].set_xlim([6.1, 7.4])
ax[1].set_xlim([6.1, 7.4])

ax[0].set_ylim([6.1, 7.4])
ax[1].set_ylim([6.1, 7.4])
```
